```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel("mlset.xlsx")
```

```python
data.head()
```

```python
data.drop(data.columns[1],axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data.iloc[:,1].str.split(' ', expand=True)
```

```python
X.head()
```

```python
X.shape
```

```python
X = X.astype(int, errors = 'raise')
```

```python
for column in X.columns:
    X[column] = X[column]  / X[column].abs().max()
```

```python
X.head()
```

```python
y=data.iloc[:,0]
```

```python
y.head()
```

```python
X_data,X_test,y_data,y_test=train_test_split(X,y,test_size=0.15,random_state=0)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X_data,y_data,test_size=0.15,random_state=0)
```

```python
pip install hypopt
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from hypopt import GridSearch
param_grid = {'C': [0.01,1,10,100,1000], 'gamma': [0.0001, 0.001,0.01,1,1.5], 'kernel': ['rbf','linear']}
opt = GridSearch(model = SVC(), param_grid = param_grid)
```

```python
opt.fit(X_train, y_train, X_val, y_val)
```

```python
print('Test Score for Optimized Parameters:', opt.score(X_test, y_test))
```

```python
param_grid = {'max_depth': [4,10,20,50]}
opt = GridSearch(model = DecisionTreeClassifier(), param_grid = param_grid)
```

```python
opt.fit(X_train, y_train, X_val, y_val)
```

```python
print('Test Score for Optimized Parameters:', opt.score(X_test, y_test))
```

```python
n=np.arange(1,10000,100)
```

```python
param_grid = {'n_neighbors': n}
opt = GridSearch(model = KNeighborsClassifier(), param_grid = param_grid)
```

```python
opt.fit(X_train, y_train, X_val, y_val)
```

```python
print('Test Score for Optimized Parameters:', opt.score(X_test, y_test))
```
